```python
from collections import Counter

# Teks untuk setiap topik
health_text = "Prioritizing health is the best investment for the future. By adopting a healthy lifestyle, our bodies become stronger and fitter, ready to face any challenges. Eating nutritious food, exercising regularly, and getting enough sleep are key to maintaining both physical and mental well-being. Maintaining personal hygiene and a clean environment are equally important for preventing illness. Prioritizing health means prioritizing a better quality of life."
sports_text = "Who said you cannot do an exercise while sitting on the office? Exercise is an activity that you can do anytime and anywhere, including in a chair where you work. Sitting too long at work can make your muscle is tense and causing muscle pain in certain parts such as the neck and back. If it continues for a long time, these habits can disrupt your health and your job.Who said you cannot do an exercise while sitting on the office? Exercise is an activity that you can do anytime and anywhere, including in a chair where you work. Sitting too long at work can make your muscle is tense and causing muscle pain in certain parts such as the neck and back. If it continues for a long time, these habits can disrupt your health and your job."
finance_text = "Financial security is one of the keys to happiness. Managing finances well can help us achieve life goals, such as buying a house, pursuing further education, or taking vacations. Creating a budget, saving money, and investing are important steps in financial planning. Avoiding an extravagant lifestyle and making wise financial decisions can help us avoid debt and achieve financial stability. With well-managed finances, we can live a more peaceful life and focus on things that matter more."

# Function to calculate BoW
def calculate_bow(text):
    # Preprocessing (optional): lowercase, stemming/lemmatization
    words = text.lower().split()  # Split words with lowercase
    bow = Counter(words)  # Count word frequencies
    return bow

# Calculate BoW for each topic
bow_health = calculate_bow(health_text)
bow_sports = calculate_bow(sports_text)
bow_finance = calculate_bow(finance_text)

# Display BoW sorted by frequency (most frequent to least frequent)
print("Health:")
for word, frequency in sorted(bow_health.items(), key=lambda item: item[1], reverse=True):
    print(f"  - {word}: {frequency}")

print("\nSports:")
for word, frequency in sorted(bow_sports.items(), key=lambda item: item[1], reverse=True):
    print(f"  - {word}: {frequency}")

print("\nFinance:")
for word, frequency in sorted(bow_finance.items(), key=lambda item: item[1], reverse=True):
    print(f"  - {word}: {frequency}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Data training
documents = [
    health_text,
    sports_text,
    finance_text
]
labels = ["health", "sports", "finance"]

# Membuat pipeline untuk vektorisasi dan klasifikasi
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Melatih model
model.fit(documents, labels)

# Data uji
test_documents = [
    "Cristiano Ronaldo came off the bench to earn Manchester United a hard-fought 2-1 victory at Everton in the Premier League on Sunday, taking his career goal tally to 700 in the process. Just as United did last weekend in their derby mauling at the hands of local rivals Manchester City, they again found themselves behind early on at Goodison Park after Alex Iwobi curled a sublime strike into the net from 20 metres."
] #teks ke-4 terjemahan dari kalimat yang ada di google classroom

# Melakukan prediksi
predictions = model.predict(test_documents)

# Menampilkan hasil prediksi
for doc, label in zip(test_documents, predictions):
    print(doc)
    print(f"Prediction: {label}")
```
